```python
# Ngõ vào là mực nước trong bồn
# Ngõ ra là điện áp cấp cho máy bơm 
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
Muc_nuoc = ctrl.Antecedent(np.arange(-1,5,0.25),'Liquid')
Voltage=ctrl.Consequent(np.arange(0,23,0.1),'Voltage')
Muc_nuoc['Low']= fuzz.trapmf(Muc_nuoc.universe,[-1,-1,0.5,1])
Muc_nuoc['moderate']=fuzz.trimf(Muc_nuoc.universe,[0.5,1,2])
Muc_nuoc['High']=fuzz.trapmf(Muc_nuoc.universe,[1,2,10,10])
Muc_nuoc.view()
Voltage['Small']= fuzz.trimf(Voltage.universe,[0,0,0])
Voltage['moderate']=fuzz.trimf(Voltage.universe,[10,10,10])
Voltage['High']=fuzz.trimf(Voltage.universe,[22,22,22])
#Voltage.view()
rule1 = ctrl.Rule(Muc_nuoc['Low'] , Voltage['High'])
rule2 = ctrl.Rule(Muc_nuoc['moderate'] , Voltage['moderate'])
rule3 = ctrl.Rule(Muc_nuoc['High'] , Voltage['Small'])
rainf_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
rainf = ctrl.ControlSystemSimulation(rainf_ctrl)
rainf.input['Liquid'] = 1.6
rainf.compute()
Voltage.view(sim=rainf)
print(rainf.output['Voltage'])
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
Fever = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Fever')
Headache = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Headache')
Nausea = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Nausea')
Vomiting = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Vomiting')
Jaundice = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Jaundice')
Liver = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Enlarge Liver')
Pain = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Joint Pain')
Body = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Body Weakness')
Dizziness = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Dizziness')
LOA = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'Loss Of Appetite')
MP = ctrl.Antecedent(np.arange(0.1,1.1,0.05),'MP')
Conclusion=ctrl.Consequent(np.arange(0.1,1.1,0.05),'Conclusion')
def ham_lien_thuoc(ham):
  ham['Mild']= fuzz.trimf(ham.universe,[0.1,0.2,0.3])
  ham['Moderate']=fuzz.trimf(ham.universe,[0.3,0.45,0.6])
  ham['Severe']=fuzz.trimf(ham.universe,[0.6,0.7,0.8])
  ham['Very_Severe']=fuzz.trimf(ham.universe,[0.8,0.9,1])
ham_lien_thuoc(Fever)
ham_lien_thuoc(Headache)
ham_lien_thuoc(Nausea)
ham_lien_thuoc(Vomiting)
ham_lien_thuoc(Jaundice)
ham_lien_thuoc(Liver)
ham_lien_thuoc(Pain)
ham_lien_thuoc(Body)
ham_lien_thuoc(Dizziness)
ham_lien_thuoc(LOA)
ham_lien_thuoc(MP)
ham_lien_thuoc(Conclusion)
Fever.view()
Conclusion.view()
a= 'Mild'
b= 'Moderate'
c= 'Severe'
d= 'Very_Severe'
rule = []
Fever_list    = [a,b,c,d,b,a,a,b,b,a,c,b,a,c,a,d,a,b,b,d]
Headache_list = [a,a,b,a,a,b,a,b,a,a,c,c,b,c,a,b,d,d,b,c]
Nausea_list   = [a,a,a,a,a,b,b,b,b,b,c,b,b,b,a,a,b,d,b,c]
Vomiting_list = [a,a,a,a,b,a,b,b,b,b,c,c,b,c,b,b,b,a,b,c]
Jaundice_list = [a,a,a,a,a,a,a,b,a,a,c,b,a,c,a,c,a,c,a,c]
Liver_list    = [a,a,a,a,a,a,a,b,c,b,c,c,a,c,a,b,b,c,a,c]
Pain_list     = [a,b,a,c,b,a,c,b,b,a,c,b,a,c,a,a,a,b,b,c]
Body_list     = [a,b,c,c,b,a,c,b,b,a,d,c,b,c,c,d,b,c,b,c]
Dizziness_list= [c,c,c,a,b,b,b,b,c,a,c,b,a,b,a,c,d,d,b,c]
LOA_list      = [a,c,c,a,c,a,b,b,b,a,c,a,b,b,c,a,a,d,b,c]
MP_list       = [a,b,b,c,b,a,b,b,b,a,b,b,b,c,b,b,b,c,b,b]
Conclusions   = [a,b,c,d,b,a,c,b,b,a,d,c,b,c,c,d,d,d,b,d]
sick_list= [Fever_list,Headache_list,Nausea_list,Vomiting_list,Jaundice_list,Liver_list,Pain_list,Body_list,Dizziness_list,LOA_list,MP_list,Conclusions]
for i in range(0,20):
    rule.append(ctrl.Rule(Fever[Fever_list[i]] & Headache[Headache_list[i]] & Nausea[Nausea_list[i]] & Vomiting[Vomiting_list[i]] & Jaundice[Jaundice_list[i]] & Liver[Liver_list[i]] & Pain[Pain_list[i]] & Body[Body_list[i]] & Dizziness[Dizziness_list[i]] & LOA[LOA_list[i]] & MP[MP_list[i]], Conclusion[Conclusions[i]] ))
# Dùng Ctrl của skfuzzy 
conclusion_control = ctrl.ControlSystem(rule)
print(conclusion_control)
conclusion_simulation = ctrl.ControlSystemSimulation(conclusion_control)
#############################################################################
conclusion_simulation.input['Fever'] =0.45
conclusion_simulation.input['Headache'] =0.5
conclusion_simulation.input['Nausea'] =0.5
conclusion_simulation.input['Vomiting'] =0.5
conclusion_simulation.input['Jaundice'] =0.5
conclusion_simulation.input['Enlarge Liver'] =0.5
conclusion_simulation.input['Joint Pain'] =0.5
conclusion_simulation.input['Body Weakness'] =0.5
conclusion_simulation.input['Dizziness'] =0.5
conclusion_simulation.input['Loss Of Appetite'] =0.5
conclusion_simulation.input['MP'] =0.5
conclusion_simulation.compute()
Conclusion.view(sim = conclusion_simulation)
```
